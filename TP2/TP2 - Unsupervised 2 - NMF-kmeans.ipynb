**Q2.**

1. NMF (Non-negative Matrix Factorization):
   - Nature of Components: NMF results in components that are non-negative, which can be useful for datasets where features are inherently non-negative, such as images.
   - Interpretability: NMF produces parts-based representations, meaning the components can represent meaningful parts of the faces, such as eyes, noses, or mouths.
   - Sparse Representation: NMF tends to produce sparse representations, where only a few components are active for any given face.
   - Visual Appearance: The eigenfaces obtained from NMF might resemble more localized facial features due to its part-based nature.

2. PCA (Principal Component Analysis):
   - Orthogonal Components: PCA generates orthogonal components, which capture the maximum variance in the data.
   - Global Structure: PCA eigenfaces tend to capture global structures and variations in lighting conditions rather than localized facial features.
   - Linear Combination: Each eigenface obtained from PCA is a linear combination of all the pixels in the original images, which can result in more smooth and blurred representations.
   - Ordered by Variance: The eigenfaces are ordered by the amount of variance they capture, with the first few eigenfaces representing the most significant variations in the data.

3. ICA (Independent Component Analysis):
   - Statistical Independence: ICA aims to find components that are statistically independent from each other, which can lead to a more disentangled representation of the data.
   - Non-Gaussianity Assumption: ICA assumes that the sources (in this case, the eigenfaces) are statistically independent and non-Gaussian, which might capture more complex patterns in the data.
   - Less Interpretable: The eigenfaces obtained from ICA might be less interpretable in terms of facial features compared to NMF or PCA, as ICA focuses on statistical independence rather than directly interpretable parts.
   - Sensitive to Mixing: ICA is sensitive to the mixing process (the way images are generated from the components), so the results might vary depending on the dataset and preprocessing steps.

In summary, while all three methods aim to extract meaningful features from face datasets, they do so in different ways, resulting in eigenfaces with distinct characteristics in terms of interpretability, sparsity, global vs. local features, and statistical assumptions.

 NMF is faster to compute than ICA due to several factors. Consequently, the computational complexity of ICA tends to be higher, leading to longer computation times, especially for large datasets or high-dimensional data like images.

 In the specific case of images of faces, NMF may present better results than PCA due to its ability to capture the inherent non-negativity and part-based structure of the data. Faces are composed of various non-negative parts such as eyes, nose, and mouth, and NMF naturally lends itself to decomposing images into such parts. This part-based representation can lead to more interpretable and meaningful components, making it easier to understand the underlying structure of the data. On the other hand, PCA, while effective at capturing global variations in the data, may struggle to represent the localized features of faces adequately. As a result, NMF can provide more expressive and discriminative features for tasks like face recognition or classification, leading to potentially better results compared to PCA in this specific context.

```python
def NNMFLecture(X, r=None, N_Iter=1000, tolerance=1e-3, plot_evolution=1):
    """
    Inputs:
        X: is a [dxN] matrix. Every column (x) is an observation and every
        row consists of features.

        r: size of the matrices W and H

        (Optional) N_Iter: maximum number of iterations

        (Optional) tolerance: convergence criteria threshold

        (Optional) plot_evolution: plot evolution convergence criteria

    Outputs:
        W: is a [d x r] matrix containing the basis images in its
        columns

        H: is a [r x N] matrix containing the loadings (h) in its columns
        of the linear combination: x=Wh
    """
    if r is None:
        r = X.shape[0]

    # Test for positive values
    if np.min(X) < 0:
        raise NameError("Input matrix X has negative values !")

    # Size
    d, N = X.shape

    # Initialization
    W = np.random.rand(d, r)
    H = np.random.rand(r, N)

    # parameters for convergence
    k = 0
    delta = np.inf
    eps = np.finfo(float).eps
    evolutionDelta = []

    while delta > tolerance and k < N_Iter:
        # multiplicative method
        XH = np.dot(X, H.T)
        HH = np.dot(H, H.T)
        for i in range(20):
            W = np.divide(W * (X @ H.T), W @ H @ H.T + eps)

        H = np.divide(H * (W.T @ X), W.T @ W @ H + eps)

        # Convergence indices
        k = k + 1
        diff = X - np.dot(W, H)
        # delta = np.sqrt(np.sum(diff**2)) / np.sqrt(np.sum(X**2)) # |X-WH|_2 / |X|_2
        # delta = sqrt(trace(diff'*diff)) / sqrt(trace(X'*X))
        delta = LA.norm(diff, "fro") / LA.norm(X, "fro")
        evolutionDelta.append(delta)

        if k == 1 or k % 100 == 0:
            print(
                "Iteration NMF number ",
                k,
                " out of ",
                N_Iter,
                ", delta = ",
                delta,
                ", error (norm delta): ",
                LA.norm(diff),
            )

    if k == N_Iter:
        print("Maximum number of iterations reached ! delta = ", delta)
    else:
        print("Convergence achieved ( delta = ", delta, ") in ", k, " iterations")

    if plot_evolution == 1:
        plt.figure(figsize=(6, 6))
        plt.plot(range(k), evolutionDelta, "bx--", linewidth=4, markersize=12)
        plt.title("Evolution of error - NMF")
        plt.show()

    return W, H
```

K-means assumes that clusters are spherical and have equal variance, which may not hold true in datasets with overlapping classes or non-spherical clusters. As a result, K-means may produce suboptimal results, failing to capture the underlying structure of the data.

3. Intersecting Classes: K-means encounters difficulties in scenarios where classes intersect or exhibit complex linear relationships with the features. K-means partitions the data based on the mean of each cluster, which may not adequately capture the intersecting nature of the classes. This can lead to misclassifications and inaccurate clustering results.

4. Concentric Class Distributions: K-means may struggle to handle datasets with concentric class distributions due to its assumption of spherical clusters. In this scenario, where the data's non-linearity challenges straightforward class prediction based on features, K-means may fail to capture the underlying structure of the data and produce suboptimal clustering results.

Furthermore, defining the correct number of clusters (k) in K-means is not always easy and can be subjective. While techniques such as the elbow method can help determine an optimal value for k, they may not always provide clear or definitive results, especially in scenarios with complex data structures or overlapping clusters. Additionally, the choice of k can significantly impact the clustering outcome, and selecting an inappropriate value may lead to erroneous interpretations of the data.

```python
## K-means
Ncluster = 2  # choose a number of clusters
kmeans = KMeans(n_clusters=Ncluster)
yKmeans = kmeans.fit_predict(X)

plt.figure(figsize=(17, 6))
plt.subplot(131)
plt.scatter(
    X[np.where(y == 0)[0], 0], X[np.where(y == 0)[0], 1], c="r", label="Class 0"
)
plt.scatter(
    X[np.where(y == 1)[0], 0], X[np.where(y == 1)[0], 1], c="b", label="Class 1"
)
plt.title("Original data")
plt.legend()

plt.subplot(132)
plt.scatter(
    X[np.where(yKmeans == 0)[0], 0],
    X[np.where(yKmeans == 0)[0], 1],
    c="r",
    label="Predicted class 0",
)
plt.scatter(
    X[np.where(yKmeans == 1)[0], 0],
    X[np.where(yKmeans == 1)[0], 1],
    c="b",
    label="Predicted class 1",
)
plt.title("K-Means")
plt.legend()

plt.subplot(133)
frontiere(kmeans, X, y, step=50)

plt.show()
```

Now, implement your own K-means using the Lloyd's algorithm.

```python
def KMeansLecture(X, K=2, tol=1e-4):
    N, d = X.shape
    delta = np.inf
    distances = np.empty((N, K))
    muTab = X[np.random.choice(N, K, replace=False)]

    while delta > tol:
        # distances[i, j] est la distance entre le point N avec le centre j
        for k in range(K):
            distances[:, k] = LA.norm((X - muTab[k]), axis=1)
        yKmeans = np.argmin(distances, axis=1)

        muTab_before = muTab.copy()

        for k in range(K):
            muTab[k] = X[yKmeans == k].mean()

        delta = LA.norm(muTab_before - muTab)

        return yKmeans, muTab
```
